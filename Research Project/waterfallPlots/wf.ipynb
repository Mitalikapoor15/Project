```python

# List of input and output file paths
input_files = [
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port1.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port2.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port3.dat"
]

output_files = [
    "port1_transposed.dat",
    "port2_transposed.dat",
    "port3_transposed.dat"
]

# Loop over files: load, transpose, and save
for in_file, out_file in zip(input_files, output_files):
    data = np.loadtxt(in_file)
    transposed = data.T
    np.savetxt(out_file, transposed, fmt="%.6f")
```

```python
# List of transposed .dat files
input_files = [
    "port1_transposed.dat",
    "port2_transposed.dat",
    "port3_transposed.dat"
]

# Waterfall plot parameters
num_rows_to_plot = 60
offset = 1
colormap_name = 'cool'

# Load and plot each file
for idx, filename in enumerate(input_files, start=1):
    # Load data
    data = np.loadtxt(filename)
    transposed = data.T

    # Create colormap
    cmap = cm.get_cmap(colormap_name, num_rows_to_plot)

    # Create waterfall plot
    plt.figure(figsize=(10, 6))
    for i in range(num_rows_to_plot):
        color = cmap(i)
        plt.plot(transposed[i] + i * offset, color=color)
        # Optional label: uncomment to enable legends
        # plt.plot(transposed[i] + i * offset, color=color, label=f"Ez {i}")

    plt.title(f"Waterfall Plot for PORT {idx}")
    plt.xlabel("Time Steps")
    plt.ylabel("Ez")
    # plt.legend()  # Optional if labels are added
    plt.grid(True)
    plt.tight_layout()

    # Save figure
    plt.savefig(f"wf_port{idx}.png", dpi=300, bbox_inches='tight')
    plt.show()
```
