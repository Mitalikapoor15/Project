```python
# List of transposed .dat files
input_files = [
    "port1_transposed.dat",
    "port2_transposed.dat",
    "port3_transposed.dat"
]

# Waterfall plot parameters
num_rows_to_plot = 60
offset = 1
colormap_name = 'cool'

# Load and plot each file
for idx, filename in enumerate(input_files, start=1):
    # Load data
    data = np.loadtxt(filename)
    transposed = data
    #not storing transposed data yet, cuz y axis needs to be the time 

    # Create colormap
    cmap = cm.get_cmap(colormap_name, num_rows_to_plot)

    # Create waterfall plot
    plt.figure(figsize=(10, 6))
    for i in range(num_rows_to_plot):
        color = cmap(i)
        plt.plot(transposed[i] + i * offset, color=color)
        # Optional label: uncomment to enable legends
        # plt.plot(transposed[i] + i * offset, color=color, label=f"Ez {i}")

    plt.title(f"Waterfall Plot for PORT {idx}")
    plt.ylabel("Time Steps")
    plt.xlabel("Ez")
    # plt.legend()  # Optional if labels are added
    plt.grid(True)
    plt.tight_layout()

    # Save figure
    plt.savefig(f"wf_port{idx}.png", dpi=300, bbox_inches='tight')
    plt.show()
    
```

```python

input_files = [
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port1.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port2.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port3.dat"
]

output_files = [
    "port1_fft_all.dat",
    "port2_fft_all.dat",
    "port3_fft_all.dat"
]

dt = 1  # Time step

for port_index, (in_file, out_file) in enumerate(zip(input_files, output_files), start=1):
    data = np.loadtxt(in_file)
    data_all = data[:, :]  # All Ez values over time

    # FFT along time axis
    fft_dat = np.abs(fft(data_all, axis=0))
    N = data_all.shape[0]
    freqs = fftfreq(N, dt)

    # Saving FFT data
    out_data = np.column_stack([freqs, fft_dat])
    np.savetxt(out_file, out_data, fmt="%.6f")

    #Waterfall
    plt.figure(figsize=(10, 6))
    offset = np.max(fft_dat) * 0.2 

    for ez_index in range(fft_dat.shape[1]):
        plt.plot(freqs, fft_dat[:, ez_index] + ez_index * offset, label=f"Ez[{ez_index}]" if ez_index < 3 else "", lw=1)
        
        
    plt.savefig(f"condensed_port {port_index}.png", dpi=300, bbox_inches='tight')
    plt.title(f"Waterfall Plot of FFT(Ez) vs Frequency - Port {port_index}")
    plt.xlabel("Frequency")
    plt.ylabel("FFT Ez Magnitude")
    plt.grid(True)
    plt.tight_layout()
    
    # plt.show()
```

```python


input_files = [
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port1.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port2.dat",
    "C:\\Users\\Mitali\\OneDrive\\Desktop\\Research Project\\portData\\port3.dat"
]

output_files = [
    "port1_fft_all.dat",
    "port2_fft_all.dat",
    "port3_fft_all.dat"
]

dt = 1  # Time step

# Choose specific Ez indices to plot
selected_ez_indices = [0, 5, 10, 15, 20]  # <-- modify this based on your data

for port_index, (in_file, out_file) in enumerate(zip(input_files, output_files), start=1):
    data = np.loadtxt(in_file)
    data_all = data[:, :]  # All Ez values over time

    # FFT along time axis
    fft_dat = np.abs(fft(data_all, axis=0))
    N = data_all.shape[0]
    freqs = fftfreq(N, dt)

    # Save FFT data
    out_data = np.column_stack([freqs, fft_dat])
    np.savetxt(out_file, out_data, fmt="%.6f")

    # Waterfall-style with selected Ez values
    plt.figure(figsize=(10, 6))
    offset = np.max(fft_dat) * 0.3  # Vertical offset between traces

    for i, ez_index in enumerate(selected_ez_indices):
        if ez_index >= fft_dat.shape[1]:
            continue  # skip if out of bounds
        plt.plot(freqs, fft_dat[:, ez_index] + i * offset, label=f"Ez[{ez_index}]", lw=1.2)

    plt.title(f"Waterfall Plot - Port {port_index}")
    plt.xlabel("Frequency")
    plt.ylabel("FFT Ez Magnitude")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
